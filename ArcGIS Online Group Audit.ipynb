# Connect to ArcGIS Online

```python
# import
from arcgis.gis import GIS
import getpass
from IPython.display import display
from arcgis.mapping import WebMap
import json
import requests
import csv
import time
from tkinter.filedialog import asksaveasfilename

# Connection Variables
Organization = input("What's your oranization? ")
User = input("What's your  username? ")
Password = getpass.getpass('Password: ')

# Connection 
gis = GIS(Organization, User, Password)
gis
```

# Group Audit Section

```python
gis = GIS(Organization, User, Password)
GroupAudit = gis.groups.search(query="", max_groups = 10000000)
s = ""
t = ""
for Group in GroupAudit:
    print("Beginning Group Audit" + "\n")
    print(str(Group) + "\n")
    GroupTitle = str(Group.get("title"))
    GroupId = str(Group.get("id"))
    GroupAccess = str(Group.get("access"))
    GroupIsFav = str(Group.get("isFav"))
    GroupIsViewOnly = str(Group.get("isViewOnly"))
    GroupisReadOnly = str(Group.get("isReadOnly"))
    GroupSortField = str(Group.get("sortField"))
    GroupTags = str(Group.get("tags"))
    ModifiedTime = str(Group.get("modified"))
    if int(ModifiedTime)<0:
        GroupModified = ""
    else:
        realtime = float(ModifiedTime[:-3])
        GroupModified = str(time.strftime("%b %d %Y %H:%M:%S", time.gmtime(realtime)))
    GroupCapabilities = str(Group.get("capabilities"))
    GroupThumbnail = str(Group.get("thumbnail"))
    GroupSnippet = str(Group.get("snippet"))
    GroupPhone = str(Group.get("phone"))
    CreatedTime = str(Group.get("created"))
    if int(CreatedTime)<0:
        GroupCreated = ""
    else:
        realtime = float(CreatedTime[:-3])
        GroupCreated = str(time.strftime("%b %d %Y %H:%M:%S", time.gmtime(realtime)))
    GroupSortOrder = str(Group.get("sortOrder"))
    GroupProvider = str(Group.get("provider"))
    GroupOwner = str(Group.get("owner"))
    GroupProviderGroupName = str(Group.get("providerGroupName"))
    GroupUsermembership = str(Group.get("userMembership"))
    GroupIsInvitationOnly = str(Group.get("isinvitationonly"))
    GroupDescription = str(Group.get("description"))
    s += GroupTitle + ","
    s += GroupId + ","
    s += GroupAccess + ","
    s += GroupIsFav + ","
    s += GroupIsViewOnly + ","
    s += GroupisReadOnly + ","
    s += GroupSortField + ","
    s += GroupTags.replace(',', "|") + ","
    s += GroupModified + ","
    s += GroupCapabilities.replace(',', "|") + ","
    s += GroupThumbnail + ","
    s += GroupSnippet.replace(',', "") + ","
    s += GroupPhone + ","
    s += GroupCreated + ","
    s += GroupSortOrder + ","
    s += GroupProvider + ","
    s += GroupOwner + ","
    s += GroupProviderGroupName + ","
    s += GroupUsermembership.replace(',', "|") + ","
    s += GroupIsInvitationOnly + ","
    GroupDescription = GroupDescription.replace('\n', " ")
    s += GroupDescription.replace(',', "") + ","
    s += "\n"
    print("End Group Audit" + "\n")
    # Beginning Group Item Audit
    GroupItem = Group.content(max_items=10000000)
    for Item in GroupItem:
        print("Beginning Group Item Audit" + "\n")
        #writelog(logFile, str(Item) + "\n")
        GroupID = GroupId
        GroupTitle = GroupTitle
        ItemID = str(Item.get("id"))
        ItemTitle = str(Item.get("title"))#.replace(',', ""))
        ItemType = str(Item.get("type"))
        ItemOwner = str(Item.get("owner"))
        CreatedTime = str(Item.get("created"))
        if int(CreatedTime)<0:
            GroupModified = ""
        else:
            realtime = float(CreatedTime[:-3])
            ItemCreated = str(time.strftime("%b %d %Y %H:%M:%S", time.gmtime(realtime)))

        ModifiedTime = str(Item.get("modified"))
        if int(ModifiedTime)<0:
            ItemModified = ""
        else:
            realtime = float(ModifiedTime[:-3])
            ItemModified = str(time.strftime("%b %d %Y %H:%M:%S", time.gmtime(realtime)))
        ItemGuid = str(Item.get("guid"))
        ItemName = str(Item.get("name"))
        ItemTypeKeywords = str(Item.get("typeKeywords"))
        ItemDescription = "" #str(Item.get("id"))
        ItemTags = str(Item.get("tags"))
        ItemSnippet = str(Item.get("snippet"))#.replace(',', "|"))
        ItemThumbnail = str(Item.get("thumbnail"))
        ItemDocumentation = str(Item.get("documentation"))
        ItemExtent = str(Item.get("extent"))
        ItemSpatialReference = str(Item.get("spatialReference"))
        ItemAccessInformation = str(Item.get("accessInformation"))
        ItemCulture = str(Item.get("culture"))
        ItemProperties = str(Item.get("properties"))
        ItemURL = str(Item.get("url"))
        ItemAccess = str(Item.get("access"))
        ItemSize = str(Item.get("size"))
        ItemAppCategories = str(Item.get("appCategories"))
        ItemIndustries = str(Item.get("industries"))
        ItemLanguages = str(Item.get("languages"))
        ItemLargeThumnail = str(Item.get("largeThumbnail"))
        ItemBanner = str(Item.get("banner"))
        ItemScreenshots = str(Item.get("screenshots"))
        itemListed = str(Item.get("listed"))
        ItemNumComments = str(Item.get("numComments"))
        ItemNumRatings = str(Item.get("numRatings"))
        ItemAvgRating = str(Item.get("avgRating"))
        ItemNumViews = str(Item.get("numViews"))
        t += GroupID + ","
        t += GroupTitle + ","
        t += ItemID + ","
        ItemTitle = ItemTitle.replace(",", ";")
        t += ItemTitle.replace('℠', "") + ","
        t += ItemType + ","
        t += ItemOwner + ","
        t += ItemCreated + ","
        t += ItemModified + ","
        t += ItemGuid + ","
        t += ItemName.replace('℠', "") + ","
        t += ItemTypeKeywords.replace(',', "|") + ","
        t += ItemDescription.replace('℠', "")  + ","
        t += ItemTags.replace(',', "|") + ","
        ItemSnippet = ItemSnippet.replace(",", ";")
        t += ItemSnippet.replace('℠', "") + ","
        t += ItemThumbnail + ","
        t += ItemDocumentation.replace('℠', "")  + ","
        t += ItemExtent.replace(',', ";") + ","
        t += ItemSpatialReference + ","
        ItemAccessInformation = ItemAccessInformation.replace('\n', " ")
        t += ItemAccessInformation.replace(',', ";") + ","
        t += ItemCulture + ","
        t += ItemProperties + ","
        t += ItemURL + ","
        t += ItemAccess + ","
        t += ItemSize + ","
        t += ItemAppCategories + ","
        t += ItemIndustries + ","
        t += ItemLanguages + ","
        t += ItemLargeThumnail + ","
        t += ItemBanner + ","
        t += ItemScreenshots.replace(',', "|") + ","
        t += itemListed + ","
        t += ItemNumComments + ","
        t += ItemNumRatings + ","
        t += ItemAvgRating + ","
        t += ItemNumViews + ","
        t += "\n"
```

# Write the Audit Results to Excel

```python
filename = asksaveasfilename() # show an "Open" dialog box and return the path to the selected file
print(filename)
AGOGroupAuditcsvfile = filename + ".csv"
AGOGroupItemAuditcsvfile = filename + "ItemAudit" + ".csv"
with open(AGOGroupAuditcsvfile, 'w') as output:
    print ("Wrtie data To CSV File")
    output.write("GroupTitle, GroupId, GroupAccess, GroupIsFav, GroupIsViewOnly, GroupisReadOnly, GroupSortField, GroupTags, GroupModified, GroupCapabilities, GroupThumbnail, GroupSnippet, GroupPhone, GroupCreated, GroupSortOrder, GroupProvider, GroupOwner, GroupProviderGroupName, GroupUsermembership, GroupIsInvitationOnly, GroupDescription\n")
    output.writelines(s)
    output2 = open(AGOGroupItemAuditcsvfile, 'w')
    output2.write("GroupID, GroupTitle, ItemID, ItemTitle, ItemType, ItemOwner, ItemCreated, ItemModified, ItemGuid, ItemName, ItemTypeKeywords, ItemDescription, ItemTags, ItemSnippet, ItemThumbnail, ItemDocumentation, ItemExtent, ItemSpatialReference, ItemAccessInformation, ItemCulture, ItemProperties, ItemURL, ItemAccess, ItemSize, ItemAppCategories, ItemIndustries, ItemLanguages, ItemLargeThumnail, ItemBanner, ItemScreenshots, itemListed, ItemNumComments, ItemNumRatings, ItemAvgRating, ItemNumViews\n")
    output2.writelines(t)   
    print("Wrtie data To CSV File Complete!")
```
